```python
# visualize the three pdf of sampling sphere/hemisphere in inversion method
from matplotlib import pyplot as plt
import random
import math

lx1 = []
ly1 = []
lz1 = []

lx2 = []
ly2 = []
lz2 = []

lx3 = []
ly3 = []
lz3 = []

fig, axs = plt.subplots(nrows=1, ncols=3, subplot_kw={"projection": "3d"}, figsize=(12, 10))

# function 1: using 1/4pi as PDF for computing CDF(theta)
for i in range(200):
    r1 = random.random()
    r2 = random.random()
    x = math.cos(2 * math.pi * r1) * 2 * math.sqrt(r2 * (1 - r2))
    y = math.sin(2 * math.pi * r1) * 2 * math.sqrt(r2 * (1 - r2))
    z = 1 - 2 * r2
    lx1.append(x)
    ly1.append(y)
    lz1.append(z)

axs[0].scatter(lx1, ly1, lz1, c='darkblue')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')
axs[0].set_zlabel('z')

# function 2: using 1/2pi as PDF for computing CDF(theta)
for i in range(200):
    r1 = random.random()
    r2 = random.random()
    x = math.cos(2 * math.pi * r1) * math.sqrt(r2 * (2 - r2))
    y = math.sin(2 * math.pi * r1) * math.sqrt(r2 * (2 - r2))
    z = 1 - r2
    lx2.append(x)
    ly2.append(y)
    lz2.append(z)

axs[1].scatter(lx2, ly2, lz2, c='darkred')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')
axs[1].set_zlabel('z')

# function 3: using cos(theta)/pi as PDF for computing CDF(theta)
for i in range(200):
    r1 = random.random()
    r2 = random.random()
    x = math.cos(2 * math.pi * r1) * math.sqrt(r2)
    y = math.sin(2 * math.pi * r1) * math.sqrt(r2)
    z = math.sqrt(1 - r2)
    lx3.append(x)
    ly3.append(y)
    lz3.append(z)

axs[2].scatter(lx3, ly3, lz3, c='darkgreen')
axs[2].set_xlabel('x')
axs[2].set_ylabel('y')
axs[2].set_zlabel('z')

plt.show()
```
